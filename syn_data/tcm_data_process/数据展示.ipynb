```python
kg_file = "TCM-KG_triples.txt"
entity_list = []
with open(kg_file, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip().split("	")
        # print(line)
        for w in line:
            w = w.strip()
            if "symmap_chemical" in w:
                continue
            if "chemical_" in w:
                continue
            if "SMIT" in w:
                continue
            
            # print(w)
            entity_list.append(w)
len(entity_list)
```

```python
def is_chinese(char):
    if '\u4e00' <= char <= '\u9fff':
        return True
    else:
        return False
chinese_entity_list = []
for i in entity_list:
    if is_chinese(i[0]) and ',' not in i:
        chinese_entity_list.append(i)
len(chinese_entity_list)
 
```
